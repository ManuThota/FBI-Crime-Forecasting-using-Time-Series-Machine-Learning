## **3. Data Wrangling**

This step focuses on cleaning and transforming the dataset to improve its quality and usability. It involves handling duplicate records, treating missing values, correcting data types, and preparing the dataset for further analysis and modeling. Proper data wrangling ensures that the model learns from accurate and consistent data.

**Handling Duplicate Values**

```python
# Removing duplicate rows from train dataset

initial_shape = train_df.shape

train_df = train_df.drop_duplicates().reset_index(drop=True)

print(f"Before Removing Duplicates: {initial_shape}")
print(f"After Removing Duplicates: {train_df.shape}")
```

**Handling Missing Values**

```python
# Checking missing values again

train_df.isnull().sum()
```

**Handling Categorical Missing Values**

```python
# Filling missing values in NEIGHBOURHOOD with 'Unknown'

train_df['NEIGHBOURHOOD'] = train_df['NEIGHBOURHOOD'].fillna('Unknown')
train_df['HUNDRED_BLOCK'] = train_df['HUNDRED_BLOCK'].fillna('Unknown')
```

**Handling Numerical Missing Values**

```python
# Filling missing values using median (robust to outliers)

num_cols = ['X', 'Y', 'Latitude', 'Longitude', 'HOUR', 'MINUTE', 'YEAR', 'MONTH', 'DAY']

for col in num_cols:
    train_df[col] = train_df[col].fillna(train_df[col].median())
```

**Handling Date Column**

```python
# Converting Date column to datetime format

train_df['Date'] = pd.to_datetime(train_df['Date'], errors='coerce')

# Reconstruct missing Date values

train_df['Date'] = train_df['Date'].fillna(
    pd.to_datetime(
        train_df[['YEAR', 'MONTH', 'DAY']].astype(int),
        errors='coerce'
    )
)

# Fill remaining missing dates with mode (most frequent date)

train_df['Date'] = train_df['Date'].fillna(train_df['Date'].mode()[0])
```

**Final Missing Values Check**

```python
# Confirm no missing values left

train_df.isnull().sum()
```

**Feature Consistency Check**

```python
# Checking data types after transformation

train_df.info()
```

**Data Cleaning & Transformation Summary with Key Insights**

- Duplicate records were identified and removed to ensure data integrity and prevent bias during model training.
- Missing values in categorical features such as **NEIGHBOURHOOD** and **HUNDRED_BLOCK** were handled by replacing them with "**Unknown**", preserving valuable data without discarding records.
- Numerical features (e.g., **HOUR, MINUTE, Latitude, Longitude**) were imputed using **median values** to maintain robustness against **outliers** and **skewed distributions**.
- The Date column was carefully reconstructed by converting it into **datetime format** and filling missing values using **YEAR, MONTH**, and **DAY**, ensuring consistency for time-based analysis.
- Remaining missing date values were handled using the **most frequent date** to maintain completeness of the dataset.
- Overall, the dataset was transformed into a **clean, consistent,** and **structured format**, making it suitable for **reliable analysis, visualization,** and **predictive modeling.**

#### **Chart 19: Time Series Decomposition**

## **6. Hypothesis Testing**

This section focuses on validating insights derived from exploratory data analysis using statistical methods. Hypothesis testing helps determine whether observed patterns in the data are statistically significant or occurred by chance. By formulating hypotheses and applying appropriate statistical tests, we aim to draw reliable conclusions that support data-driven decision-making and strengthen the overall credibility of the analysis.


#### **Hypothetical Statement - 1**

"Crime occurrences significantly vary across different hours of the day."

**Step 1: Define Hypotheses**
> Null Hypothesis (H₀):
> - Crime occurrences are uniformly distributed across all hours (no significant difference)
>
> Alternative Hypothesis (H₁):
> - Crime occurrences vary significantly across different hours

**Step 2: Statistical Test**

> Chi-Square Goodness of Fit Test
>
> Why?
> - Data is categorical (hours)
> - We are comparing observed vs expected distribution

```python
# Hypothesis Test 1: Crime distribution across hours

from scipy.stats import chisquare

# Observed frequency
observed = train_df['HOUR'].value_counts().sort_index()

# Expected frequency (uniform distribution)
expected = [observed.mean()] * len(observed)

# Perform Chi-Square test
chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("Chi-Square Statistic:", chi_stat)
print("P-value:", p_value)
```

**Interpretation**

```python
alpha = 0.05

if p_value < alpha:
    print("Reject Null Hypothesis → Crime varies across hours")
else:
    print("Fail to Reject Null Hypothesis → No significant variation")
```

**Hypothesis 1 — Result & Conclusion**
> Test Outcome:
> - Chi-Square Statistic: 78465.47
> - P-value: 0.0

**Interpretation:**

> Since the p-value (0.0) is less than the significance level (α = 0.05), we reject the null hypothesis.

**This means:**

> Crime occurrences are not uniformly distributed across hours and vary significantly depending on the time of day.

**Key Insight:**
> - Crime is time-dependent, not random
> - Certain hours (especially evening/night) experience higher crime frequency
> - Lower activity during early morning hours

**Business Impact:**
> Optimized Patrol Scheduling:
> - Police resources can be deployed more heavily during peak crime hours

**Risk if Ignored:**

**Final Conclusion:**

The analysis confirms that time plays a critical role in crime occurrence, and incorporating time-based strategies can significantly improve public safety and operational efficiency.

#### **7.2 Handling Outliers**

**Outlier Detection Code**
> First, we visualize outliers using boxplots

```python
# Visualizing outliers using boxplots

num_cols = ['X', 'Y', 'Latitude', 'Longitude', 'HOUR', 'MINUTE']

plt.figure(figsize=(15,8))

for i, col in enumerate(num_cols, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(y=train_df[col])
    plt.title(col)

plt.tight_layout()
plt.show()
```

**Outlier Treatment Code (IQR Method)**

```python
# Handling outliers using IQR method

for col in num_cols:
    Q1 = train_df[col].quantile(0.25)
    Q3 = train_df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Capping the outliers
    train_df[col] = np.where(train_df[col] < lower_bound, lower_bound, train_df[col])
    train_df[col] = np.where(train_df[col] > upper_bound, upper_bound, train_df[col])
```

**Handling Outliers — Explanation**
- Boxplot visualizations revealed the presence of outliers in several numerical features, particularly in **spatial variables** such as **X, Y, Latitude, and Longitude**.
- Extreme values were also observed in some temporal features, although **HOUR and MINUTE** largely followed expected ranges with minimal abnormal behavior.
- The presence of outliers in location-based features may be due to:
    - Data entry inconsistencies
    - Measurement errors
    - Rare but valid extreme geographical points
- To handle these outliers, the **Interquartile Range (IQR)** method was applied.
Values outside the range defined by (Q1 - 1.5 * IQR) and (Q3 + 1.5 * IQR) were identified as **outliers**.
- Instead of removing these data points, **capping (winsorization)** was performed, where extreme values were replaced with the calculated **lower and upper bounds**.

**Why this technique was used:**
- Preserves data (no row deletion)
- Prevents model distortion due to extreme values
- Robust and widely used statistical method
- Suitable for real-world datasets with natural variability

#### **ML Model 1: ARIMA**

**ARIMA (AutoRegressive Integrated Moving Average)** is a classical time series forecasting model.

It uses:
- AR (AutoRegressive) → past values
- I (Integrated) → differencing (to make data stationary)
- MA (Moving Average) → past errors

ARIMA is suitable for:
- Univariate time series
- Capturing trends in data

**Import & Train Model**

```python
from statsmodels.tsa.arima.model import ARIMA
# To Suppress the warnings

import warnings
warnings.filterwarnings("ignore")

# Use stationary data (differenced)
model_arima = ARIMA(train_ts['Total_Crimes_diff'], order=(1,1,1))

arima_model = model_arima.fit()

print(arima_model.summary())

```

**Cross Validation**

```python
# cross validation

model_cv = SARIMAX(
    train_ts['Total_Crimes_diff'],
    order=(1,0,1),
    seasonal_order=(1,1,1,12)
).fit()

forecast_cv = model_cv.forecast(steps=len(test_ts))

cv_score_sarima = r2_score(test_ts['Total_Crimes_diff'], forecast_cv)

print("CV R2 Score:", cv_score_sarima)
```

**Hyperparameter Tuning**

```python
# imprting for tuning
import itertools

# Implementing Hyperparameter tuning
p = q = range(0, 2)
d = [0]

seasonal_pdq = [(x[0], x[1], x[2], 12) for x in itertools.product(p, d, q)]

best_score = float("inf")
best_params = None

for param in itertools.product(p, d, q):
    for seasonal in seasonal_pdq:
        try:
            model = SARIMAX(
                train_ts['Total_Crimes_diff'],
                order=param,
                seasonal_order=seasonal
            ).fit()

            forecast = model.forecast(steps=len(test_ts))
            rmse = np.sqrt(mean_squared_error(test_ts['Total_Crimes_diff'], forecast))

            if rmse < best_score:
                best_score = rmse
                best_params = (param, seasonal)

        except:
            continue
# Printing the Best records

print("Best Params:", best_params)
print("Best RMSE:", best_score)
```
